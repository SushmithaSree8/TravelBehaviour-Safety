# Data Extraction

Subsets the required variables from each of the 5 datasets (Identification of Sample Households, Household Characteristics, Demographics, Expenditure on Miscellaneaous, Expenditure on Purchase and Construction 

```python
# import statements
import pandas as pd
import numpy as np
```

### Identification of Sample Household

```python
# reading orginal file
df_1=pd.read_csv('Identification of Sample Household.txt', delimiter='\t', header=None, encoding = 'utf8')
df_1.head()
```

```python
# subsetting columns required
df_1_req=df_1[[0, 7, 8, 9, 22, 35]]
df_1_req.columns=['hh_id','state','region','district','response_code','multiplier']
df_1_req.head()
```

```python
#saving as csv file
df_1_req.to_csv('IdentificationOfSampleHousehold.csv',index=False)
```

### Household Characteristics

```python
# reading orginal file
df_2=pd.read_csv('Household Characteristics.txt', delimiter='\t', header=None, encoding = 'utf8')
df_2.head()
```

```python
# subsetting columns required
df_2_req=df_2[[0,21,22,23,24,25,26,36,37,38,39,42,43,44]]
df_2_req.columns=['hh_id','hh_size','nic_2004','nco_2004','hh_type','religion','social_group',
                'cooking_code','lighting_code','dwelling_unit','salary_income','internet_access','mpce_urp','mpce_mrp']
df_2_req.head()
```

```python
#saving as csv file
df_2_req.to_csv('HouseholdCharacteristics.csv',index=False)
```

### Demographic and other particulars of household members

```python
# reading orginal file
df_3=pd.read_csv('Demographic and other particulars of household members.txt', delimiter='\t', header=None, encoding = 'utf8')
df_3.head()
```

```python
# subsetting columns required
df_3_req=df_3.iloc[:,np.r_[0, 21:34]]
df_3_req.columns=['hh_id','person_serial_no','relation','sex','age',
                 'marital_status', 'education','days_stayed_away','meals_per_day','meals_school',
                 'meals_employer','meals_others','meals_payment','meals_home']
df_3_req.head()
```

```python
#saving as csv file
df_3_req.to_csv('Demographics.csv',index=False)
```

### Expenditure on miscellaneous goods and services

```python
# reading orginal file
df_4=pd.read_csv('Expenditure on miscellaneous goods and services including medical (non-institutional), rents and taxes during the last 30 days.txt', delimiter='\t', header=None, encoding = 'utf8')
df_4.head()
```

```python
# subsetting columns required
df_4_req=df_4[[0, 21, 22]]
df_4_req.columns=['hh_id','item_code','value']
df_4_req.head()
```

```python
#saving as csv file
df_4_req.to_csv('ExpenditureOnMiscellaneousGoods.csv',index=False)
```

### Expenditure for purchase and construction

```python
# reading orginal file
df_5=pd.read_csv('Expenditure for purchase and construction (including repair and maintenance) of durable goods for domestic use.txt', delimiter='\t', header=None, encoding = 'utf8')
df_5.head()
```

```python
# subsetting columns required
df_5_req=df_5.iloc[:,np.r_[0, 21:36]]
df_5_req.columns=['hh_id','item_code','whether_possess','first_hand_purchase_no_30days','whether_hire_purchased_30days',
             'first_hand_purchase_value_30days','cost_raw_material_repair_30days','sec_hand_purchase_value_30days','total_expenditure_30days',
             'first_hand_purchase_no_365days','whether_hire_purchased_365days','first_hand_purchase_value_365days','cost_raw_material_repair_365days',
             'sec_hand_purchase_no_365days','sec_hand_purchase_value_365days','total_expenditure_365days']
df_5_req.head()
```

```python
#saving as csv file
df_5_req.to_csv('ExpenditureForPurchaseAndConstruction.csv',index=False)
```
